```python
# preprocessing , 4 category 
df['insurance_validity'].value_counts()
```

```python
d1 = {'Comprehensive':0,'Third Party insurance':1,
                              'Zero Dep':2,'Not Available':3,'Third Party':1}
df['insurance_validity']=df['insurance_validity'].map(d1)
```

```python
df['insurance_validity'].value_counts()
```

```python
# 2
df['fuel_type'].value_counts()
```

```python
d2 = {'Petrol':0,'Diesel':1,'CNG':2}
df['fuel_type']=df['fuel_type'].map(d2)
```

```python
df['fuel_type'].value_counts()
```

```python
# 3
df['ownsership'].value_counts()
```

```python
d4 = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fifth Owner':5,'Fourth Owner':4}
df['ownsership']=df['ownsership'].map(d4)
```

```python
df['ownsership'].value_counts()
```

```python
# 4
df['transmission'].value_counts()
```

```python
d3= {'Manual':0,'Automatic':1}
df['transmission']=df['transmission'].map(d3)
```

```python
df['transmission'].value_counts()
```

```python
df['registration_year'].value_counts()
```

```python
df['registration_year'] = df['registration_year'].astype(str).str[-2:].astype(int)
df['registration_year']
```

```python
X = df[['insurance_validity','fuel_type','kms_driven','ownsership','transmission','registration_year']]
y = df['price(in lakhs)']
```

```python
X.head()
```

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

```python
from sklearn.linear_model import LinearRegression 
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
models={'LinReg':LinearRegression()
        ,'KNN-5':KNeighborsRegressor(n_neighbors=5),
        'SVM-Poly':SVR(kernel='poly')} 
# R2 Score ecould be negative and model is worse
for name, model in models.items():
    model.fit(X_scaled, y)
    sc = model.score(X_scaled, y)
    print(name, sc)
```

```python
final_model=models['KNN-5']
```

```python
# input function always in string 
insurance_validity= input('insurance_validity:')
fuel_type= input('fuel_type:')
kms_driven= input('kms_driven:')
ownsership= input('ownsership:')
transmission= input('transmission:')
registration_year = input('registration_year:')

insurance_validity = d1[insurance_validity]
fuel_type = d2[fuel_type]
transmission = d3[transmission]
ownsership = d4[ownsership]
kms_driven =int(kms_driven)
registration_year = int(str(registration_year)[-2:])


test = [[insurance_validity, fuel_type, kms_driven, ownsership, transmission, registration_year]]
test_scaled = scaler.transform(test)
yp = final_model.predict(test_scaled)[0]
print(f'Predicted Price: {yp:.2f} lakhs')
print(f'Predicted Price in ₹: {yp*100000:.0f}')
```

```python
import pickle
pickle.dump(final_model,open('final_model.pkl','wb'))
pickle.dump(scaler, open('scaler.pkl','wb'))
```
